```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
```

```python
try:
    # Assuming the file is in the same directory as this notebook
    df = pd.read_csv('sms_spam_perceptron.csv')
except FileNotFoundError:
    print("Error: The file 'sms_spam_perceptron.csv' was not found. Please check the file path.")
```

```python
y = df.iloc[:, 0].values
y = np.where(y == 'spam', -1, 1)
x = df.iloc[:, [1, 2]].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
p = Perceptron(max_iter=40, eta0=0.1, random_state=0)
p.fit(x_train, y_train)
```

```python
y_pred = p.predict(x_test)
```

```python
misclassified = (y_test != y_pred).sum()
accuracy = accuracy_score(y_test, y_pred)
```

```python
print(f"Misclassified samples: {misclassified}")
print(f"Accuracy: {accuracy:.2f}")
```
